Histogram displaying normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt 

values=np.random.randint(0,11,10000)
histogram=np.bincount(values)
h=histogram/histogram.sum()
print(h)

plt.figure(figsize=(10,6))
plt.title("Frequency of Values")
plt.xlabel("Values")
plt.ylabel("Frequency")
counts,bin_edges,patches=plt.hist(values,color="#D31A48")
for i in range(len(counts)):
    x=(bin_edges[i]+bin_edges[i+1])/2
    y=counts[i]
    plt.text(x,y,str(y),ha='center',va='bottom',fontsize=8,color="#3A23C0")
plt.show()
```

Poisson Distribution

```python
#Average rate=5
poisson=np.random.poisson(5,10)
count=np.bincount(poisson)
print(count)
print(poisson)
```

Central Law Theorem

```python
M = 10000
m = np.zeros(M)
for i in range(M):
    t = np.random.beta(5,2,size=M)
    m[i] = t.mean()
print(m)
```

Law of Large Numbers

```python
m = []
for n in np.linspace(1,8,30):
    t = np.random.normal(1,1,size=int(10**n))
    m.append(float(t.mean()))
print(m)
```
