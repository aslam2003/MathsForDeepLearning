Variance

```python
import numpy as np
N=100000000
data=np.random.randint(0,4000,N)
biased=((data-data.mean())**2).sum()/N
unbiased=((data-data.mean())**2).sum()/(N-1)
print(biased)
print(unbiased)
```

Quartiles

```python
scores=np.random.randint(0,101,1000)
q=np.quantile(scores,q=[0.0,0.25,0.5,0.75,1.0])
print(q)
print(len(scores[((q[0]<=scores) & (scores<q[1]))]))
print(len(scores[((q[1]<=scores) & (scores<q[2]))]))
print(len(scores[((q[2]<=scores) & (scores<q[3]))]))
print(len(scores[((q[3]<=scores) & (scores<=q[4]))]))
```

Missing Values->Mean and Median

```python
data=np.random.randint(0,1001,1000)
data=data.astype("float")

for i in range(50):
    idx=np.random.randint(0,1000)
    data[idx]=np.nan

copy1=data.copy()
copy2=data.copy()

total=copy1[~(np.isnan(copy1))].sum()
non_value_counts=len(copy1[~(np.isnan(copy1))])
mean=np.round(total/non_value_counts,2)
print(mean)

copy1=np.where(np.isnan(copy1),mean,copy1)
print(copy1.mean())

scores=copy2[~(np.isnan(copy2))]
median=np.median(scores)
print(median)

copy2=np.where(np.isnan(copy2),median,copy2)
print(np.median(copy2))
```

Boxplot

```python
import matplotlib.pyplot as plt
import numpy as np

# Simulate F1 scores from 10-Fold Cross Validation
# Model A: Consistent performance (Low Variance)
model_a_scores = [0.84, 0.85, 0.83, 0.85, 0.86, 0.84, 0.85, 0.82, 0.85, 0.84]

# Model B: Unpredictable performance (High Variance)
# Note: The average is similar to A, but the spread is huge.
model_b_scores = [0.70, 0.95, 0.65, 0.92, 0.85, 0.75, 0.98, 0.60, 0.88, 0.82]

# Create the Boxplot
plt.figure(figsize=(8, 5))
plt.boxplot([model_a_scores, model_b_scores], labels=['Model A (Stable)', 'Model B (Unstable)'])

plt.title('Stability Comparison: 10-Fold CV F1 Scores')
plt.ylabel('F1 Score')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
