Broadcasting

```python
import numpy as np

#Broadcasting conditions
#->Equal rows
#->Equal columns

A=np.array([[1,2,3],[4,5,6]])
B=np.array([4,5,8])
C=np.array([2,3])

print(np.add(A,B))
#print(np.add(A,C))->No equal rows or columns

#Reshaping C for broadcasting
C=C.reshape((2,1))
print(np.add(A,C))
```

Projection of Vectors

```python
a=np.array([3,4,9])
b=np.array([2,-5,-4])

#Projection of vector A onto B
p=(np.dot(a,b)/np.dot(b,b))*b
print(p)
```

Cross,Outer,Kronecker Product

```python
v1=np.array([3,9,12])
v2=np.array([19,-2,0])

m1=np.array([[3,4,5,8],[4,5,9,3]])
m2=np.array([[13,14,25,32],[90,23,41,20]])

cross=np.cross(v1,v2)
outer=np.outer(m1,m2)
kronecker=np.kron(m1,m2)

print("Cross Product: ",cross)
print("Outer Product: ",outer)
print("Kronecker Product: ",kronecker)
```
