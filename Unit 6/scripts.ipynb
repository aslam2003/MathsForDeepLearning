Trace,Diagonals,Matrix Power

```python
import numpy as np

matrix_A=np.array([[3,4,5],[9,2,7],[13,12,15]])
trace=np.trace(matrix_A)
diagonal=np.diagonal(matrix_A)

#A^5,A^10
matrix_A5=np.linalg.matrix_power(matrix_A,5)
matrix_A10=np.linalg.matrix_power(matrix_A,10)

print("Trace:",trace)
print("Diagonal",diagonal)
print("A^5",matrix_A5)
print("A^10",matrix_A10)
```

Upper Triangular,Lower Triangular

```python
upper_triangular=np.triu(matrix_A)
lower_triangular=np.tril(matrix_A)

print("Upper :",upper_triangular)
print("Lower: ",lower_triangular)
```
