Monte Carlo Simulation

```python
import numpy as np 
N=10000000 #Iterations
M=3 # 3 coin flips
heads=np.zeros(M+1)

for i in range(N):
    flips=np.random.randint(0,2,M)
    h,t=np.bincount(flips,minlength=2)
    heads[h]+=1

prob=heads/N
print("Probability:%s",np.array(prob).astype(str))
```

Birthday Paradox

```python
#Finding number of comparisons where people sharing a birthday is 99%
def nC2(n):
    answer=n*(n-1)/2
    return answer/2

res=0    
for i in range(2000):
    if((364/365)**i<0.01):
        print(i)
        res=i
        break

for i in range(10000):
    if nC2(i)>=res:
        print(i)
        break


```

```python
#Probability two people sharing a birthday in 1000000
match=0
for i in range(100000):
    a=np.random.randint(0,365)
    b=np.random.randint(0,365)
    if a==b:
        match+=1
print(f"Probability of a match: {match/1000000:0.6f}")
```

```python
#Probability for m people to share a birthday in 1000000

def findprob(m):
    matches=0
    for i in range(1000000):
        match=0
        b=np.random.randint(0,365,m)
        for i in range(m):
            for j in range(i+1,m):
                if i!=j and b[i]==b[j]:
                    match+=1
        if match>0:
            matches+=1
    
    print("Probabity:",matches/1000000)

findprob(10)
```

```python
#Probability of picking 4 people out of 50 people who haven't been to boston
#Assume people with ID 0,1,2,3, have been to boston
#We test it over N=1000000 simulations

N=1000000
picks=0
for i in range(N):
    people=np.random.randint(0,50,3)
    fail=False
    for t in range(3):
        if people[t]<4:
            fail=True
    if(not fail):
        picks+=1

print("Probability:",picks/N)
```
